#### Requirement 1: 
  Write a SQL query that will combine the Creative_Data and Performance_Data, for ads with "tesla" in the Ad_Name. Also include an additional metric called "Purchases_Efficiency" which should be Link_Clicks / Purchases


```python
'''
SELECT
    Creative_Data.*, 
    Performance_Data.*,
    ROUND( 
        ( CAST(Performance_Data.Link_Clicks AS DECIMAL(5,2)) / CAST(Performance_Data.Purchases as DECIMAL(5,2)) )
        , 2 
        ) AS Purchases_Efficiency
    FROM Creative_Data
    INNER JOIN Performance_Data
    ON Creative_Data.id = Performance_Data.user_id
    WHERE email LIKE '%tesla%'
    ORDER BY Purchases_Efficiency
    ;
'''
```

## 2.
    "Download" the files from the server by creating a Pandas Dataframe table of both the Creative_Data and Performance_Data (call it "Server_df, and ignore the need to filter for "tesla" above)

```python
creative = pd.read_csv('./Creative_data.csv', index_col=False)
creative.head()
```

```python
creative.shape
```

```python
performance = pd.read_csv('./Performance_data.csv', index_col=False)
performance.head()
```

```python
performance.shape
```

```python
corrected = pd.read_csv('./Corrected_Creative_data.csv', index_col=False)
corrected.head()
```

```python
corrected.shape
```

```python
(creative.columns == corrected.columns).all()
```

```python
server_df = creative.merge(performance, how='inner', indicator=False, on='Ad_ID', validate='one_to_one')
server_df.head()
```

```python
server_df.shape
```

## 3.
    Compare the corrected file with the Dataframe you just created - output a new Dataframe with only the differences (call it "Differences_df)

```python
differences_df = server_df.merge(corrected, how = 'right' ,indicator=True).loc[lambda x : x['_merge']=='right_only'].drop(columns='_merge')
differences_df.head()
```

```python
differences_df.shape
```

## 4.
    Update Server_df with the corrected values (don't just use the file with the correct tags as the server dataframe, please amend the original Server_df from step #2)

```python
server_df.update(differences_df)
```
